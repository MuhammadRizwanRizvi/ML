k-means clustering is an un-supervised learning, it predicts the result when there 
is no already labels for data, it includes dividing data into k clusters and 
perform the calculations to predict the cluster of new data.
    for this centroids are to be choosen, now distance from points to each center
will be calculated and the center having smallest value of distance will be the 
cluster of new points, then the centriod is modified by taking mean of previous and new points.

```python
#make clusters of students using their gpa and iq
import pandas as pd
```

```python
df=pd.read_csv('clustering.csv')
df.tail(5)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

```python
#there are fours clusters 
x=df.iloc[:,:].values
x.shape
```

```python
import sklearn as sk
sk.__version__
```

```python
from sklearn.cluster import KMeans
#fitting the model for 4 clusters
model=KMeans(n_clusters=4)
model.fit(x)
model.predict(x)
```

```python
labels=model.labels_
centers=model.cluster_centers_
print("cluster numbers :\n",labels,"\ncentriods of clusters:\n",centers)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=labels,label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1],marker='x', c='red',label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('iq-level')
plt.ylabel('cgpa')
plt.legend()
plt.grid(True)
plt.show()
```
