```python
#praticing matplotlib
#matplotlib is a data visvualizing library, it allows to plot data in graphs,bars,pie bars, histograms and scatter graph.
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_csv('insurance.csv')
df.head(3)
```

```python
#first parameter to .plot is x-axis value and 2nd parametr is y-axis value
#basic plot
plt.plot(df['bmi'],df['charges'])
plt.title('simple plotting')
plt.show()
```

```python
import numpy as np
x=np.array([10,20])
y=np.array([5,10])
plt.title("plotting two points only, not a whole line",loc='left')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#if x-axis value is not given the by defualt it will take 0,1,2.. depend on y-axis value
plt.plot(x,y,'o')
plt.show()
```

```python
x=np.array([5,10,15,20])
y=np.array([1,2,3,4])
#implementing marker , color,size,style
plt.plot(x,y,'o--r',ms=20,mec='blue')
plt.show()
```

```python
#applying grid on graph
plt.plot(x,y,'o:g',ms=10,mec='yellow')
plt.grid(color='green')
```

```python
#subplots
#single row 2 columns plot number 1
plt.subplot(1,2,1)
plt.plot(x,y)
#plot number 2
plt.subplot(1,2,2)
plt.plot(x*2,y*2)
plt.suptitle("subplots")
```

```python
plt.scatter(df['bmi'],df['charges'])
plt.title('single scatter plot')
plt.show()
```

```python
#green shows first plot
plt.scatter(df['bmi'],df['charges'],color='green')
plt.title('combine scatter plots')
#pink shows 2nd plot
plt.scatter(df['age'],df['charges'],color='pink')
#showing color bar
plt.colorbar()
plt.show()
```

```python
plt.title('practicing matplotlib bars')
plt.bar(df['smoker'],df['charges'],color='pink',width=0.1)
plt.show()
```

```python
#histograms
plt.title('trying matplotlib histograms')
plt.hist(df['charges'],color='grey')
plt.show()
```

```python
#pie charts
plt.title('matplotlib pie bars')
plt.pie(df['charges'])
plt.show()
```
