```python
import pandas as pd
```

panda is a machine learning library which help to make the data frame for training

```python
#reading .csv file
df=pd.read_csv("rawdata.csv")
df
```

```python
#checking either any column has null entry or not, ture if there is any null and false if there is no null entry
df.isnull()
```

```python
#dropping the null entries
df.dropna(inplace=True,axis=0)
df
```

```python
df.dropna(axis=1,inplace=True)
df
```

```python
#filling null values with 123
dfnew=pd.read_csv("rawdata.csv")
dfnew.fillna(value=123,inplace=True)
dfnew
```

```python
#forward filling of null entries, it means fill row 1 has value 5 and row 2 is empty then fill 5 in row 2 also
dfnew=pd.read_csv("rawdata.csv")
dfnew.fillna(method='ffill',inplace=True)
dfnew
```

```python
#backward filling, if row 2 is 5 and row 1 is empty then fill row 1 with 5
dfnew=pd.read_csv("rawdata.csv")
dfnew.fillna(method='bfill',inplace=True)
dfnew
```

```python
#filling using quadratic value of that column
dfnew=pd.read_csv("rawdata.csv")
dfnew.interpolate(method='quadratic',inplace=True)
dfnew
```

```python
#changing type of column
dfnew['Calories']=dfnew['Calories'].astype(float)
dfnew['Calories']
```

```python
dfnew['Calories'].apply(lambda x: x+2)
```

```python
#check if there is any duplicte or not , true if there is any duplicate
dfnew.duplicated()
```

```python
#dropping if there any duplicates
dfnew.drop_duplicates(subset=['Calories'])
```
